1. Выведите количество пустых значений по колонкам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во пустых значений)

```python
pd.read_sql(
    '''
    SELECT 'CRIM' as 'column', count()
    FROM boston
    where CRIM is null
    UNION
    SELECT 'ZN' as 'column' , count()
    FROM boston
    where ZN is null
    UNION
    SELECT 'INDUS' as 'column' , count()
    FROM boston
    where INDUS is null
    UNION
    SELECT 'CHAS' as 'column' , count()
    FROM boston
    where CHAS is null
    UNION
    SELECT 'NOX' as 'column' , count()
    FROM boston
    where NOX is null
    ''',
    con,
)
```

2. Выведите количество уникальных значений по колонокам CRIM, ZN, INDUS, CHAS, NOX (название колонки, кол-во уникальных значений)

```python
pd.read_sql(
    '''
    SELECT 'CRIM' as column_name, count(distinct CRIM) as cnt_distinct
    FROM boston
    union all
    SELECT 'ZN' as column_name, count(distinct ZN) as cnt_distinct
    FROM boston
    union all
    SELECT 'INDUS' as column_name, count(distinct INDUS) as cnt_distinct
    FROM boston
    union all
    SELECT 'CHAS' as column_name, count(distinct CHAS) as cnt_distinct
    FROM boston
    union all
    SELECT 'NOX' as column_name, count(distinct NOX) as cnt_distinct
    FROM boston
    ''',
    con,
)
```

3. Выведите колонки, у которых медиана равна минимальному значению (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX.
Напишите какой вывод можно сделать по данным в этих колонках

```python
pd.read_sql(
    '''
    select 'CRIM' as column_name
    from (
    select min,
    case when prepare_for_median = '1_even' then (sum(CRIM) over())/2 else sum(CRIM) over() end as median
    from (
    SELECT CRIM,
    min(CRIM) over() as min,
    row_number() over(order by CRIM) as row_num,
    count() over() as cnt_all,
    case when (count() over())%2 = 0 then '1_even' else
    case when (count() over())%2 != 0 then '2_odd' end end as prepare_for_median
    FROM boston ) t
    where case when prepare_for_median = '1_even' then t.row_num = ((cnt_all/2)+1) or t.row_num = cnt_all/2 else t.row_num = cnt_all/2 end) t2
    where min = median
  union
    select 'ZN' as column_name
    from (
    select min,
    case when prepare_for_median = '1_even' then (sum(ZN) over())/2 else sum(ZN) over() end as median
    from (
    SELECT ZN,
    min(ZN) over() as min,
    row_number() over(order by ZN) as row_num,
    count() over() as cnt_all,
    case when (count() over())%2 = 0 then '1_even' else
    case when (count() over())%2 != 0 then '2_odd' end end as prepare_for_median
    FROM boston ) t
    where case when prepare_for_median = '1_even' then t.row_num = ((cnt_all/2)+1) or t.row_num = cnt_all/2 else t.row_num = cnt_all/2 end) t2
    where min = median
  union
    select 'INDUS' as column_name
    from (
    select min,
    case when prepare_for_median = '1_even' then (sum(INDUS) over())/2 else sum(INDUS) over() end as median
    from (
    SELECT INDUS,
    min(INDUS) over() as min,
    row_number() over(order by INDUS) as row_num,
    count() over() as cnt_all,
    case when (count() over())%2 = 0 then '1_even' else
    case when (count() over())%2 != 0 then '2_odd' end end as prepare_for_median
    FROM boston ) t
    where case when prepare_for_median = '1_even' then t.row_num = ((cnt_all/2)+1) or t.row_num = cnt_all/2 else t.row_num = cnt_all/2 end) t2
    where min = median
  union
    select 'CHAS' as column_name
    from (
    select min,
    case when prepare_for_median = '1_even' then (sum(CHAS) over())/2 else sum(CHAS) over() end as median
    from (
    SELECT CHAS,
    min(CHAS) over() as min,
    row_number() over(order by CHAS) as row_num,
    count() over() as cnt_all,
    case when (count() over())%2 = 0 then '1_even' else
    case when (count() over())%2 != 0 then '2_odd' end end as prepare_for_median
    FROM boston ) t
    where case when prepare_for_median = '1_even' then t.row_num = ((cnt_all/2)+1) or t.row_num = cnt_all/2 else t.row_num = cnt_all/2 end) t2
    where min = median
  union
    select 'NOX' as column_name
    from (
    select min,
    case when prepare_for_median = '1_even' then (sum(NOX) over())/2 else sum(NOX) over() end as median
    from (
    SELECT NOX,
    min(NOX) over() as min,
    row_number() over(order by NOX) as row_num,
    count() over() as cnt_all,
    case when (count() over())%2 = 0 then '1_even' else
    case when (count() over())%2 != 0 then '2_odd' end end as prepare_for_median
    FROM boston ) t
    where case when prepare_for_median = '1_even' then t.row_num = ((cnt_all/2)+1) or t.row_num = cnt_all/2 else t.row_num = cnt_all/2 end) t2
    where min = median
    ''',
    con,
)
```

Можно сделать вывод, что распределение величины в этих колонках скошено. В отсортированных данных преобладает минимальное значение, из-за чего медиана, которая отражает 50% процентиль в данных, оказывается равной минимальному значению.

4. Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг) среди всех значений LSTAT. Напишите какой вывод можно сделать по этим данным.

```python

pd.read_sql(
    '''
   select t.LSTAT, t.MEDV, t.rank_num  from (
    SELECT max(MEDV) over() as max_medv, rank() over(order by LSTAT) as rank_num,
    *
    FROM boston ) t
    where t.max_medv = t.medv
    ''',
    con,
)
```

В домах с самой дорогой стоимостью процент населения с низким статусом не превышает 10%.

5. Выведите среднюю стоимость домов граничащих с рекой(CHAS) и нет (граничит/не граничит, стоимость)

```python

pd.read_sql(
    '''
    SELECT distinct case when CHAS = 0.0 then 'не граничит' else 'граничит' end as river_border,
    avg(MEDV) over(partition by CHAS) as avg_price
    FROM boston
    ''',
    con,
)
```

6. Выведите все колонки, у которых среднее значение выше, когда дом граничит с рекой (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX. Напишите какой вывод можно сделать по этим данным.

```python

pd.read_sql(
    '''
    select 'CRIM' as column_name from (
    select *, lead(avg) over() as avg_river_border, row_number() over() as rn
    from (
    SELECT distinct round(avg(CRIM) over(),2) as avg
    FROM boston
    UNION
    SELECT distinct round(avg(CRIM) over(),2) as avg
    FROM boston
    where chas = 1.0 ) t ) t1
    where t1.rn = 1 and avg_river_border > avg
UNION
    select 'ZN' as column_name from (
    select *, lead(avg) over() as avg_river_border, row_number() over() as rn
    from (
    SELECT distinct round(avg(ZN) over(),2) as avg
    FROM boston
    UNION
    SELECT distinct round(avg(ZN) over(),2) as avg
    FROM boston
    where chas = 1.0 ) t ) t1
    where t1.rn = 1 and avg_river_border > avg
UNION
    select 'INDUS' as column_name from (
    select *, lead(avg) over() as avg_river_border, row_number() over() as rn
    from (
    SELECT distinct round(avg(INDUS) over(),2) as avg
    FROM boston
    UNION
    SELECT distinct round(avg(INDUS) over(),2) as avg
    FROM boston
    where chas = 1.0 ) t ) t1
    where t1.rn = 1 and avg_river_border > avg
UNION
    select 'CHAS' as column_name from (
    select *, lead(avg) over() as avg_river_border, row_number() over() as rn
    from (
    SELECT distinct round(avg(CHAS) over(),2) as avg
    FROM boston
    UNION
    SELECT distinct round(avg(CHAS) over(),2) as avg
    FROM boston
    where chas = 1.0 ) t ) t1
    where t1.rn = 1 and avg_river_border > avg
UNION
    select 'NOX' as column_name from (
    select *, lead(avg) over() as avg_river_border, row_number() over() as rn
    from (
    SELECT distinct round(avg(NOX) over(),2) as avg
    FROM boston
    UNION
    SELECT distinct round(avg(NOX) over(),2) as avg
    FROM boston
    where chas = 1.0 ) t ) t1
    where t1.rn = 1 and avg_river_border > avg
    ''',
    con,
)
```

Вывод: граница с рекой однозначно повышает среднее значение указанных переменных.

7. Выведите значения долей промышленной застройки(INDUS), концентрации оксидов азота(NOX) и по их перцентилям - 10, 20 ... 100 ( перцетиль(10,20...100),значение INDUS, значение NOX). Напишите прослеживается между ними взаимосвязь
